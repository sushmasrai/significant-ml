### Detecting Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def get_vif_factors( X ):
    X_matrix = X.as_matrix()
    vif = [ variance_inflation_factor( X_matrix, i ) for i in range( X_matrix.shape[1] ) ]
    vif_factors = pd.DataFrame()
    vif_factors['column'] = X.columns
    vif_factors['vif'] = vif
    
    return vif_factors
```

```python
## TODO: Get VIF Factors for the X Variables
vif_factors = get_vif_factors(X[X_features])
```

### Residual Analysis

### P-P plot

```python
def draw_pp_plot( model ):
    probplot = smf.ProbPlot( model.resid )
    plt.figure( figsize = (8, 6) )
    probplot.ppplot( line='45' )
    plt.title( "Normal P-P Plot of Regression Standardized Residuals" )
    plt.show()
```

```python
## TODO: Draw pp plot for the model 2
draw_pp_plot(ipl_model_2)
```

```python
def get_standardized_values( vals ):    
    return (vals - vals.mean())/vals.std()
```

```python
def plot_resid_fitted( fitted, resid):
    plt.scatter( get_standardized_values( fitted ), 
            get_standardized_values( resid ) )
    plt.title( "Residual Plot" )
    plt.show()
```
